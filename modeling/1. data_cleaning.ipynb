```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df_customer = pd.read_csv('../dataset/customer.csv')
df_customer.head()
```

```python
df_customer.info()
```

```python
df_customer.isnull().sum()
```

```python
#fill null value with modus
df_customer['Marital Status'] = df_customer['Marital Status'].fillna(df_customer['Marital Status'].mode()[0])
df_customer.isnull().sum()
```

```python
df_customer.describe().T
```

```python
df_product = pd.read_csv('../dataset/product.csv')
df_product.head()
```

```python
df_product.info()
```

```python
df_product.isnull().sum()
```

```python
df_product.describe().T
```

```python
df_store = pd.read_csv('../dataset/store.csv')
df_store.head()
```

```python
df_store.info()
```

```python
df_store.isnull().sum()
```

```python
df_store.describe().T
```

```python
df_transaction = pd.read_csv('../dataset/transaction.csv')
df_transaction.head()
```

```python
df_transaction.info()
```

```python
df_transaction.isnull().sum()
```

```python
df_transaction.describe().T
```

```python
#convert Date to datetime
df_transaction['Date'] = pd.to_datetime(df_transaction['Date'],format='mixed')
```

```python
# merge dataset

df_merge = pd.merge(df_transaction,df_product, on='ProductID', how='left')
df_merge = pd.merge(df_merge,df_store, on='StoreID', how='left')
df_merge = pd.merge(df_merge,df_customer, on='CustomerID', how='left')
df_merge.head()
```

```python
df_drop = df_merge.copy()
df_drop.drop(['TransactionID','ProductID','Latitude','Longitude','CustomerID','StoreID'],axis=1,inplace=True)
df_drop.head()
```

```python
df_drop.info()
```

```python
df_drop.isnull().sum()
```

```python
#define the feature
cat = df_drop.select_dtypes(include='object').columns
num = df_drop.select_dtypes(exclude='object').columns

print(f'There are {len(cat)} categorical features and {len(num)} numerical features')
print(f'the categorical features are {cat}')
print(f'the numerical features are {num}')
```

```python
#show distribution of numerical features
plt.figure(figsize=(16,9))
for i in range(len(num)):
    plt.subplot(3,3,i+1)
    sns.distplot(df_drop[num[i]])
    plt.tight_layout()
```

```python
#show distribution of categorical features
plt.figure(figsize=(16,9))
for i in range(len(cat)):
    plt.subplot(3,3,i+1)
    sns.countplot(x=df_drop[cat[i]])
    plt.tight_layout()
    plt.xticks(rotation=90)
```

```python
#save df_drop into csv file
df_drop.to_csv('../dataset/data.csv',index=False)
```
